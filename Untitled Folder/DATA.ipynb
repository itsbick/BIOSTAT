# Using  .loc[]


More advanced slicing of dataframes can be done via .loc[] and .iloc[], .loc[] is discussed first


```python
# loc[] accepts two things within its bracket, separate by a,
# though technically, it can accept only 1 agument, say a number (when such is the case, it donates a  row)
# on the lift are the indices (number) of the rows (as a range ) and the right, column name/s enclosed as a list
# the range used by the rows is a little different from the way we use rows though, as loc rows are inclusive
# say from example, I want the first 10 rows of the sepal.length column


# also just to prove I have 10 
print(iris.loc[0:9,['sepal.length']].shape)
```

```python
iris.loc[0:9,['sepal.length']]
```

```python
# if one wants to get the columns via loc[]
iris.loc[:,['sepal.length']]
```

```python
 
# if to get 1 row only, say, row 7 (index 6)
iris.loc[6:6]
```

```python
# though this one is simpler to type, but  the output isn’t the usual row one might think
iris.loc[6]
```

```python
# iris.loc[6,] can be done too!
iris.loc[6,]
```

```python
# the start or end of the range can be omitted, kinda like this
# say, from first row to 5th
iris.loc[:4,]
```

```python
# or the 140th row until the end for petal.width and variety
iris.loc[140:, ['petal.width', 'variety']]
```

# Slicing via column values


Dataframes can also be sliced on column values (think of this as fulfilling a certain condition, which can yield only True or False values). An example:


```python
iris.variety.value_counts()
```

Base  on the code cell above, the dataframe iris, on its variety columns, contains 3 values (Setoss, Versicolor, Virnica, each appearing 50 times). Say, one wants to slice the dataframe thar removes all varieties except Versicolor, how is that done?


```python
# see first that the condition is either True or False
print(list(iris['variety'] == 'Versicolor'))
```

```python
iris['variety'] == 'Versicolor'
```

```python
# then index those that are True
iris[iris['variety'] == 'Versicolor']
```

Or how about keeping all Versicolor whose patal width is greater than 1.4?


```python
iris[(iris['variety'] == 'Versicolor') & (iris['petal.width'] > 1.4)]
     
# take note, for multiple conditions, use “&”
# “and” is used for something else
# check the output: there’s no petal width 1.4 and below
 
```

# Aggregating functions


```python
# suppose one wants to get the mean, median and sum altogether for sepal length
# aggregate all functions via the aggregate method
# pasa a dictionary as argument, with the column as key and the functions as values
irisgroups.agg({'sepal.length' :['mean', 'sum']})
```

```python
irisgroups.agg({'sepal.length' :['mean', 'median', 'sum'], 'sepal.width' :['mean', 'median', 'sum']})
```
